Fraud Detection Menggunakan Metode Logistic Regression

1. Load Data dan Import Libraries

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import pandas as pd
path="/content/drive/MyDrive/dataset/fraud_detection_bank_dataset.csv"
df = pd.read_csv(path)
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2. Praproses Data

```python
#Membagi Data
X = df.iloc[:,1:-1].values
Y = df.iloc[:,-1].values
```

```python
#Normalisasi Data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
#Split Train dan Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=10)
```

3. Model Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(X_train, Y_train)
```

4. Hasil

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test,y_pred)
precision = precision_score(Y_test,y_pred)
recall = recall_score(Y_test,y_pred)
f1 = f1_score(Y_test,y_pred)
```

```python
print("Accuracy :",accuracy)
print("Precision :",precision)
print("Recall :",recall)
print("F1 Score :",f1)
```

5. Dengan Reduksi Dimensi Menggunakan PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=11)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents,columns = ['pc 1','pc 2','pc 3','pc 4','pc 5','pc 6','pc 7','pc 8','pc 9','pc 10','pc 11'])
principalDf
```

```python
finalDf = pd.concat([principalDf, df[['targets']]], axis = 1)
finalDf
```

```python
#Split Train dan Test Data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(principalDf, df[['targets']], test_size = 0.2, random_state=10)

#Model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(X_train, Y_train)

#Hasil
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
y_pred = model.predict(X_test)
accuracy = accuracy_score(Y_test,y_pred)
precision = precision_score(Y_test,y_pred)
recall = recall_score(Y_test,y_pred)
f1 = f1_score(Y_test,y_pred)
print("Accuracy :",accuracy)
print("Precision :",precision)
print("Recall :",recall)
print("F1 Score :",f1)
```
